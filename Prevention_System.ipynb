```python
df = pd.read_csv("Identifying-Suicidal-Ideation-from-Twitter---A-Detection-and-Prevention-System/Prevention_dataset.csv")
```

```python
df.head()
```

```python
t= df['date'].to_numpy()
```

```python
import datetime
from datetime import datetime
from datetime import time
```

```python
dates=[]
for i in t:
   i=i.replace("PDT","")
   dates.append(datetime.strptime(i,'%a %b %d %H:%M:%S %Y'))
```

```python
dates
```

```python
df['date']= dates
```

```python
times = []
for date in dates:
  times.append(date.time())
```

```python
df['times']= times
times
```

```python
nparray=df[['user','times']].to_numpy()
nparray
```

```python
u= df.user.unique()
len(u)
```

```python
cols=['Username','Freq(11pm-6am)','Freq(6am-11pm)','Total Freq']
```

```python
data=pd.DataFrame(columns=cols)
```

```python
data['Username']=u
```

```python
data.head()
```

```python
freq_1=np.zeros(len(u))
freq_2=np.zeros(len(u))
count=0
ll= time(23,00,00)
ul= time(6,0,0)
for user in u:
  for item in nparray:
    if(user==item[0]):
      if(((item[1]>=ll) and (item[1]<=time(23,59,59))) or ((item[1]>=time(0,0,0)) and (item[1]<=ul))):
        freq_1[count]+=1
      else:
        freq_2[count]+=1
  count+=1
```

```python
data['Freq(11pm-6am)']=freq_1
data['Freq(6am-11pm)']=freq_2
```

```python
data['Total Freq']=data['Freq(11pm-6am)']+data['Freq(6am-11pm)']
```

```python
data
```

```python
data.to_csv('Identifying-Suicidal-Ideation-from-Twitter---A-Detection-and-Prevention-System/Frequency Data.csv',index=False)
```
